```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv(r"E:\Code_R\health_iFEMG_US\Data\iFEMG\iFEMG_curl_all.csv", header=0, index_col=None)
data_df.head()
```

```python
# 提取特征，标准化
x = data_df.iloc[:, 6:30]
x_scaled = StandardScaler().fit_transform(x)
print(f"数据集维度：{x.shape}")

normalised_breast = pd.DataFrame(x_scaled, columns=list(x))

# PCA
pca_model = PCA(n_components=3)
principalComponents_breast = pca_model.fit_transform(x_scaled)
# 建立主成分df
principal_breast_Df = pd.DataFrame(data = principalComponents_breast,
                                   columns = ['principal component 1', 'principal component 2', 'principal component 3'])
print('Explained variation per principal component: {}'.format(pca_model.explained_variance_ratio_))
```
