```python
"""
批量读取.csv特征集(bicps or tricps)
进行数据合并、归一化等操作
"""
import pandas as pd

from utils_iFEMG_feature import df_save_csv, df_merge_colname
```

```python
# 健康人数据集
# 读取文件路径下的文件
file_df_list = []

folder_path = r"E:\Data\paper2\iFEMG"
subject_name_list = ['hpy', 'lmh', 'lmt', 'lpy-1', 'lpy-2', 'pym-1', 'pym-2', 'wcx', 'zjz', 'zk-1', 'zk-2', 'zpk-1', 'zpk-2', 'zx']

for name in subject_name_list:
    file_path = folder_path + '\\' + name + '\\' + name + "_extension_abs.csv"
    try:
        file_df_list.append(pd.read_csv(file_path, header = [0,1], index_col = None))
    except FileNotFoundError:
        print(f"文件 {file_path} 未读取")
    pass

print(f"文件数: {len(file_df_list)}")
```

```python
# 积水潭患者数据集
# 读取文件路径下的文件
file_df_list = []

folder_path = r"E:\Data\paper2\积水潭患者数据集"
subject_name_list = ['w8s1', '0912s1', '0912s2', '0919s2', '0926s1', '0926s2', '1010s1', '1017s1', '1017s2', '1024s1', '1107s1']

for name in subject_name_list:
    file_path = folder_path + '\\' + name + '\\affected\\' + name + "_curl_abs.csv"
    try:
        file_df_list.append(pd.read_csv(file_path, header = [0,1], index_col = None))
    except FileNotFoundError:
        print(f"文件 {file_path} 未读取")
    pass

print(f"文件数: {len(file_df_list)}")
```

```python
# 列表中的df合并
final_data_df = pd.concat(file_df_list, axis = 0, ignore_index = True)
final_data_df_n = df_merge_colname(final_data_df)
final_data_df_n
```
