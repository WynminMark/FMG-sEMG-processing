```python
'''
计算一次实验中的多路iFEMG信号的特征

更改4通道iFEMG采集板
1600采样率

没有计算主动与拮抗肌肉之间的差值'''

import pandas as pd

import sys
sys.path.append('..')
from utils.utils_iFEMG_feature import df_save_csv
from utils.utils_gui_model import form_sbj_info_df, form_feature_df_1ch
```

```python
"""
初始化一些全局变量，只需运行一次
"""
feature_name_lst = ['initial_pressure_ave', 'FMG', 'mav', 'rms', 'wave_length', 'zero_crossing', 'slope_sign_change', 'mean_freq', 'mean_power_freq']

# 文件名及其对应的key
file_name_dict = {'aqian0':['affected', 'qian', '0'],
                  'aqian1_5':['affected', 'qian', '1.5'],
                  'aqian3':['affected', 'qian', '3'],
                  'ahou0':['affected', 'hou', '0'],
                  'ahou1_5':['affected', 'hou', '1.5'],
                  'ahou3':['affected', 'hou', '3'],
                  'hqian0':['healthy', 'qian', '0'],
                  'hqian1_5':['healthy', 'qian', '1.5'],
                  'hqian3':['healthy', 'qian', '3'],
                  'hhou0':['healthy', 'hou', '0'],
                  'hhou1_5':['healthy', 'hou', '1.5'],
                  'hhou3':['healthy', 'hou', '3']
                  }
# 列名与肌肉名称，仅在传感器通道与上述肌肉位置不对应时需要更改
b1_col_names_lst = ['time', 'FMG_RF', 'FMG_LF', 'FMG_SarT', 'FMG_BF', 'sEMG_RF', 'sEMG_LF', 'sEMG_SarT', 'sEMG_BF']
b2_col_names_lst = ['time', 'FMG_SemiT', 'FMG_VM', 'sEMG_SemiT', 'sEMG_VM']

# 肌肉名称
muscle_name_lst = ['RF', 'LF', 'SarT', 'BF', 'SemiT', 'VM']
```

```python
# 计算特征集
folder_path = r"E:\fuckinggggg\paper_iFEMG_4th\data"
# 241114s1, 241128s2, 241204s3, 241204s4, 241206s5, 241206s6
subject_name = "241206s6"
# sample rate
fs = 1610   # 1626 or 1222
channel_nums = 4

###########################################################################
# 以上都需要改
############################################################################

# 用于存储每个db文件计算的特征
temp_df_list = []

for key, value in file_name_dict.items():
    # ch1 RF
    temp_fea_df1 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 1,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs, 
                                    ch_nums=channel_nums)
    if temp_fea_df1 is not None:
        temp_fea_df1.columns = ['RF' + i for i in feature_name_lst]
    else:
        print(f"跳过当前循环")
        continue    # 如果为空则跳过当前循环，不产生temp_df

    # ch2 LF
    temp_fea_df2 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 2,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs,
                                    ch_nums=channel_nums)
    if temp_fea_df2 is not None:
        temp_fea_df2.columns = ['LF' + i for i in feature_name_lst]
        pass

    # ch3 SarT
    temp_fea_df3 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 3,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs,
                                    ch_nums=channel_nums)
    if temp_fea_df3 is not None:
        temp_fea_df3.columns = ['SarT' + i for i in feature_name_lst]
        pass

    # ch4 BF
    temp_fea_df4 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 4,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs,
                                    ch_nums=channel_nums)
    if temp_fea_df4 is not None:
        temp_fea_df4.columns = ['BF' + i for i in feature_name_lst]
        pass

    # ch5 SemiT
    temp_fea_df5 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b2' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 1,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs,
                                    ch_nums=channel_nums)
    if temp_fea_df5 is not None:
        temp_fea_df5.columns = ['SemiT' + i for i in feature_name_lst]
        pass

    # ch6 VM
    temp_fea_df6 = form_feature_df_1ch(db_file_path = folder_path + '\\' + subject_name + '\\' + 'b2' + '\\' + key + '.db',
                                    time_file_path = folder_path + '\\' + subject_name + '\\' + 'b1' + '\\' + key + '.txt',
                                    signal_channel = 2,
                                    abandon_ms = 1000,
                                    signal_sample_freq = fs,
                                    ch_nums=channel_nums)
    if temp_fea_df6 is not None:
        temp_fea_df6.columns = ['VM' + i for i in feature_name_lst]
        pass

    # 合并df
    temp_df = pd.concat([temp_fea_df1, temp_fea_df2, temp_fea_df3, temp_fea_df4, temp_fea_df5, temp_fea_df6], axis=1)

    # 添加受试者个人信息和重量label
    temp_df['subject'] = subject_name
    temp_df['side'] = value[0]
    temp_df['motion_type'] = value[1]
    temp_df['label'] = value[2]
    
    print(f"特征数量：{temp_df.shape}")
    temp_df_list.append(temp_df)
    # all_feature_df = pd.concat([all_feature_df, temp_df], axis = 0, ignore_index=True)
    # print(all_feature_df.shape)
    pass

all_feature_df = pd.concat(temp_df_list, axis = 0, ignore_index=True)
print(all_feature_df.shape)
all_feature_df.head
```
