```python
"""
批量读取.csv特征集(bicps or tricps)
对每个人的数据进行单独归一化再merge
"""
import pandas as pd

from utils_iFEMG_feature import df_save_csv, df_norm, df_merge_colname
```

```python
"""
获取文件路径下多个文件夹中.csv格式的特征集

1. 文件夹路径
2. 肌肉bi tri
"""
# 文件夹路径
folder_path = r"E:\Data\paper2\iFEMG"
subject_name_list = ['hpy', 'lmh', 'lmt', 'lpy-1', 'lpy-2', 'pym-1', 'pym-2', 'wcx', 'zjz', 'zk-1', 'zk-2', 'zpk-1', 'zpk-2', 'zx']


file_df_list = []
for name in subject_name_list:
    file_path = folder_path + '\\' + name + '\\' + name + "_extension_abs.csv"
    try:
        file_df_list.append(pd.read_csv(file_path, header = [0,1], index_col = None))
    except FileNotFoundError:
        print(f"文件 {file_path} 未读取")
    pass

print(f"文件数: {len(file_df_list)}")
```

```python
"""
对每名被试信号特征分别单独归一化

NOTE:
    - 需要修改保存文件的文件名！！
"""

# 归一化某特征
normed_df_list = [] 
err_df_index = 0    # 如果有key error发生报错，报告所在df的index
for df in file_df_list:
    # 对指定的列进行归一化
    try:
        """
        normed_df_list.append(df_norm(dataframe=df, col_name=[("bicps_br", "act_ave_list"),
                                                            ("tricps_br_medial", "act_ave_list"), 
                                                            ("tricps_br_lateral", "act_ave_list")], method='min-max'))
                                                            
                    'FMG', 'mav', 'rms', 'wave_length', 'zero_crossing', 'slope_sign_change', 'mean_freq', 'mean_power_freq'
                                                """
        normed_df_list.append(df_norm(dataframe=df, col_name=[("bicps_br", "initial_pressure_ave"),
                                                              ("bicps_br", "FMG"),
                                                              ("bicps_br", "mav"),
                                                              ("bicps_br", "rms"),
                                                              ("bicps_br", "wave_length"),
                                                              ("bicps_br", "zero_crossing"),
                                                              ("bicps_br", "slope_sign_change"),
                                                              ("bicps_br", "mean_freq"),
                                                              ("bicps_br", "mean_power_freq"),
                                                              ("tricps_br_medial", "initial_pressure_ave"),
                                                              ("tricps_br_medial", "FMG"),
                                                              ("tricps_br_medial", "mav"),
                                                              ("tricps_br_medial", "rms"),
                                                              ("tricps_br_medial", "wave_length"),
                                                              ("tricps_br_medial", "zero_crossing"),
                                                              ("tricps_br_medial", "slope_sign_change"),
                                                              ("tricps_br_medial", "mean_freq"),
                                                              ("tricps_br_medial", "mean_power_freq"),
                                                              ("tricps_br_lateral", "initial_pressure_ave"),
                                                              ("tricps_br_lateral", "FMG"),
                                                              ("tricps_br_lateral", "mav"),
                                                              ("tricps_br_lateral", "rms"),
                                                              ("tricps_br_lateral", "wave_length"),
                                                              ("tricps_br_lateral", "zero_crossing"),
                                                              ("tricps_br_lateral", "slope_sign_change"),
                                                              ("tricps_br_lateral", "mean_freq"),
                                                              ("tricps_br_lateral", "mean_power_freq")], method='z_score'))
    except KeyError:
        print(f"index {err_df_index} 存在错误")
        pass
    err_df_index = err_df_index + 1
    pass
# 合并dataframe
final_data_df = pd.concat(normed_df_list, axis = 0, ignore_index = True)
```

```python
# 合并列名
final_data_df_n = df_merge_colname(final_data_df)
# 更改列名
final_data_df_n.rename(columns={'subject_info_subject_name': 'subject_info_height',
                                'subject_info_height': 'subject_info_weight',
                                'subject_info_weight': 'subject_info_age',
                                'subject_info_age': 'subject_info_gender',
                                'subject_info_gender': 'subject_info_name',}, inplace=True)
final_data_df_n
```

```python
# 保存文件
df_save_csv(final_data_df_n, r"E:\Data\paper2\iFEMG\iFEMG_extension_abs_normed_all.csv")
```
